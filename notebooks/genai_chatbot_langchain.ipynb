## Experiments on building chatbot with LangChain and LlamaIndex

LlamaIndex is mainly a data framework for RAG systems, while LangChain is an application framework for building end-to-end LLM workflows.

Chat history is created and passed to llm to have full context. Chat history holds messages.
Each message will have a role and content. There are three roles - system, user and assistant.

In LangChain, prompt template is used for building the prompt dynamically. Out parser is used to convert the raw output from LLM to string format. 


# Simple chatbot with LangChain

```python
! pip install -q langchain-community langchain-groq
```

```python
import os
from langchain_core.prompts import ChatPromptTemplate 
from langchain_core.output_parsers import StrOutputParser 
from langchain_groq import ChatGroq # Can use ollama or open  ai or anything
from dotenv import load_dotenv
load_dotenv()
api_key = os.environ["GROQ_API_KEY"]

llm = ChatGroq(
    api_key = api_key,
    model="llama-3.3-70b-versatile",
    temperature = 0
)
parser = StrOutputParser()
```

```python
# Build prompt with chat history, create chain with prompt, llm and parser, invoke the chain
def chat_langchain():
    print("Welcome to LangChain Chatbot. To quit, type 'exit' \n")
    chat_history = [
        ("system", "Your a helpful chatbot. Be accurate and concise while answering")
    ]
    while True:
        user_input = input("You: ").strip()
        if user_input.lower() == "exit":
            print("\nThank you. Have a nice day!")
            break
        chat_history.append(("user", user_input))

        prompt = ChatPromptTemplate.from_messages(chat_history) # Dynamic building of prompt

        chain = prompt | llm | parser

        response = chain.invoke({}) # Get the response

        print(f"Chatbot: {response}\n")

        chat_history.append(("assistant", response))

        print("-"*80)

chat_langchain()
```

# Simple chatbot with LlamaIndex

```python
! pip install -q llama-index llama-index-llms-groq
```

```python
import os
from llama_index.core.llms import ChatMessage, MessageRole
from llama_index.llms.groq import Groq
from dotenv import load_dotenv

load_dotenv()
api_key=os.environ["GROQ_API_KEY"]

llm = Groq(
    api_key = api_key,
    model =  "llama-3.3-70b-versatile",
    temperature = 0
)
```

```python
def chat_llamaindex():
    print("Welcome to LlamaIndex Chatbot. To quit, type 'exit' \n")
    chat_history = [
        ChatMessage(role=MessageRole.SYSTEM, content="You are a helpful assistant. Be accurate and concise while answering")
    ]
    while True:
        user_input = input("You: ").strip()
        if user_input.lower() == "exit":
            print("\nThank you. Have a nice day!")
            break
        chat_history.append(ChatMessage(role=MessageRole.USER, content=user_input))

        response = llm.chat(messages=chat_history)
        answer = response.message.content

        print(f"Chatbot: {answer}\n")

        chat_history.append(ChatMessage(role=MessageRole.ASSISTANT, content=answer))
        print("-"*80)

chat_llamaindex()
```
