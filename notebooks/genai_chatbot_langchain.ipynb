## Experiments on building chatbot with LangChain 

Prompt template is used for building the prompt dynamically. Out parser is used to conver the raw output from LLM to string format. 
Chat history is created and passed to llm to have full context. Chat history is a list of messages.
Each message will have a role and content. There are three roles - system, user and assistant.

```python
import os
from langchain_core.prompts import ChatPromptTemplate 
from langchain_core.output_parsers import StrOutputParser 
from langchain_groq import ChatGroq # Can use ollama or open  ai or anything
from dotenv import load_dotenv
load_dotenv()
api_key = os.environ["GROQ_API_KEY"]

llm = ChatGroq(
    api_key = api_key,
    model="llama-3.3-70b-versatile",
    temperature = 0
)
parser = StrOutputParser()
```

```python
# Build prompt with chat history, create chain with prompt, llm and parser, invoke the chain
def chat():
    print("Welcome to LangChain Chatbot. To quit, type 'exit' \n")
    chat_history = [
        ("system", "Your a helpful chatbot. Be accurate and concise while answering")
    ]
    while True:
        user_input = input("You: ").strip()
        if user_input.lower() == "exit":
            print("\nThank you. Have a nice day!")
            break
        chat_history.append(("user", user_input))

        prompt = ChatPromptTemplate.from_messages(chat_history) # Dynamic building of prompt

        chain = prompt | llm | parser

        response = chain.invoke({}) # Get the response

        print(f"Chatbot: {response}\n")

        chat_history.append(("assistant", response))

        print("-"*80)

chat()
```
